```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\mkows\OneDrive\Desktop\New folder\Bank_Personal_Loan_Modelling.csv")
print("Dataset length:",len(data))
print("Dataset Shape:",data.shape)

print("Top 5 data in Dataset:")
data.head(5)
```

```python
X=data.drop('CreditCard',axis=1)
y=data['CreditCard']

print(X.shape)
print(y.shape)
```

```python
sns.scatterplot(x='Experience',y='CCAvg',data=data)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10,test_size=0.15)
```

```python
clf_gini=DecisionTreeClassifier(criterion='gini',random_state=10,max_depth=3,min_samples_leaf=5)
clf_gini.fit(X_train,y_train)
```

```python
y_pred_gini=clf_gini.predict(X_test)
print("Results Using Gini Index:")
print("Predicted values:")
print(y_pred_gini)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_gini))
print("Accuracy Score:")
dt_gini=accuracy_score(y_test,y_pred_gini)*100
dt_gini
```

```python
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=10,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train,y_train)
```

```python
y_pred_entropy=clf_entropy.predict(X_test)
print("Result Using Entropy:")
print("predicted Values:")
print(y_pred_entropy)
```

```python
print("Confusion matrix:\n  \n \n",confusion_matrix(y_test,y_pred_entropy))
```

```python
print("accuracy score:")
dt_entropy=accuracy_score(y_test,y_pred_entropy)*100
dt_entropy
```

```python
print("f1 score Report:\n",classification_report(y_test,y_pred_entropy))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\mkows\OneDrive\Desktop\New folder\Bank_Personal_Loan_Modelling.csv")
print(data)
```

```python
X=data.iloc[:,0:14].values
print(X.shape)
print(X[:5])
```

```python
data[:5]
```

```python
y=data.iloc[:,13].values
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
print("--------------------")
print(X_train[:5])

print(X_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

random=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
random.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

random1=RandomForestClassifier(n_estimators=10,criterion='gini',random_state=0)
random1.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
ypre=random.predict(X_test)
ypre1=random1.predict(X_test)
```

```python
print("Entropy Accuracy Score:")
rf_accuracy_entropy=accuracy_score(y_test,ypre)*100
rf_accuracy_entropy
```

```python
print('gini Accuracy Score:')
rf_accuracy_gini=accuracy_score(y_test,ypre1)*100
rf_accuracy_gini
```

```python
confusion_matrix(y_test,ypre1)
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, ypre)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, ypre1)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
